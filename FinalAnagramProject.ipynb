###### **Loading and sorting the word list given**

```python
wordlist = sorted(list(set([word.strip().lower() for word in open('words.txt','r')])))
```

###### **Storing the Anagrams in a dictionary one word from each group is the key**

```python
anaDic={}


for text in wordlist:
    simple_text = ''.join(sorted(text.lower().strip()))
    if simple_text in anaDic:
        anaDic[simple_text].append(text)
    else:
        anaDic[simple_text]=[text]
    
    
```

######  Grouping the anagrams into a list

```python
anaGroups=[]

for text in anaDic:
    if len(anaDic[text])>1:
        anaGroups.append(anaDic[text])
```

###### Creating the final Dictionary storing the word length as the key and the anagram group lists as values

```python
finalAnagrams={}

for groups in anaGroups:
    
    if len(groups[0]) in finalAnagrams:
          finalAnagrams[len(groups[0])].append(groups)
    else:
          finalAnagrams[len(groups[0])]=[groups]
    
```
